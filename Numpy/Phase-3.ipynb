```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array(
    [
        [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub
        [4, 180000, 210000, 240000, 270000],  # Burger Point
        [5, 160000, 185000, 205000, 230000],  # Chai Point
    ]
)

print("==== Zomato Data Analysis ====")
print("\nSales data shape : ", sales_data.shape)
print("\n Sample Data for 1st three restaurants:\n ", sales_data[0:3])
```

```python
# Total Sales per year

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("Total Sales per year: \n", yearly_total)

# minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum sales per restaurant: \n", min_sales)

# maximum sales per year
max_sales_year = np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales per year: \n", max_sales_year)

# Average Sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales : \n", avg_sales)

# Cumalative sales
cum_sum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sum: \n ", cum_sum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cum_sum, axis=0))
plt.title("Average Cumulative Sales across all Restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector add = ", vector1 + vector2)  # matrix Addition
print("\n Multiplication = ", vector1 * vector2)  # Matrix Multiplication, same rules
print("Dot product of Vector = ", np.dot(vector1, vector2))
angle = np.arccos(
    np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
)
print("Angle between two vectors = ", angle)
```
