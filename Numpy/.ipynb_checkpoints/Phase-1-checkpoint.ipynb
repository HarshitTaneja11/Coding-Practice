## Learning Numpy


```python
import numpy as np
```

#### Creating Array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array: ")
print(arr_2d)
```

# List vs Numpy Array


```python
py_list = [1, 2, 3]
print("Python List Multiplication ", py_list * 2)

np_array = np.array([1, 2, 3])  # element wise multiplication
print("Numpy array: ", np_array * 10)
```

```python
import time
import numpy as np

start = time.time()
py_list = [i * 2 for i in range(10000000)]
print("\n List operation time = ", time.time() - start)

start = time.time()
np_array = np.arange(10000000) * 2
print("\n Numpy operation time = ", time.time() - start)
```

# Creating Array from scratch


```python
zeros = np.zeros((3, 4))  # gives zero matrix array with 3 row and 4 col
print("Zeros array: \n", zeros)

ones = np.ones((2, 3))  # 2x3 Matrix array with 1 as all elements
print("Ones array: \n", ones)

full = np.full((2, 2), 7)  # 2x2 Array with 7 in all spaces
print("Full Array: \n", full)

random = np.random.random((4, 3))
print("Random Array: \n", random)

sequence = np.arange(0, 10, 2)  # start, stop, step
print("Sequence Array: \n", sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3, 4, 5])
print("Vector: ", vector)
print()

matrix = np.array([[1, 2, 3], [4, 5, 6]])  # Matrix is 2d array
print("Matrix: \n", matrix)
print()

Tensor = np.array(
    [
        [[1, 2], [3, 4]],  # Tensor has more dimensions, multidimensional array
        [[5, 6], [7, 8]],
    ]
)
print("Tensor: \n", Tensor)
print()
```

# Array Properties

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Shape: ", arr.shape)  # tells how many rows and columns are there
print("Dimension: ", arr.ndim)  # Tells dimension
print("Size: ", arr.size)  # Gives total number of elements
print("Data Type: ", arr.dtype)
```

### Array Reshaping


```python
arr = np.arange(12)
print("Orignal Array ", arr)

reshaped_arr = arr.reshape((3, 4))  # converts in 3x4 matrix
print("Reshaped Array \n", reshaped_arr)

flattened_arr = reshaped_arr.flatten()  # Flats into single row
print("Flattened Array \n", flattened_arr)

raveled = reshaped_arr.ravel()  # returns view instead of copy
print("Raveled Array \n", raveled)

transpose_arr = reshaped_arr.transpose()  # rows -> column and vice versa
print("Transpose Array \n", transpose_arr)
```
